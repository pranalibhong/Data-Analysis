Distribution of Monthly Income Among Employees With Different Job Roles

```python

plt.figure(figsize=(6, 5))
sns.boxplot(x='JobRole', y='MonthlyIncome', data=data)
plt.title('Distribution of Monthly Income Among Employees With Different Job Roles')
plt.xlabel('Job Role')
plt.ylabel('Monthly Income')
plt.xticks(rotation=45)
plt.show()
```

Effect of Distance From Home on Employee Attrition

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Attrition', y='DistanceFromHome', data=data)
plt.title('Effect of Distance From Home on Employee Attrition')
plt.xlabel('Attrition (No = 0, Yes = 1)')
plt.ylabel('Distance From Home')
plt.show()

# Group by 'Attrition' and calculate the mean distance from home
average_distance = data.groupby('Attrition')['DistanceFromHome'].mean()
print(average_distance)
```

# Distribution of Employees' Ages Across Different Departments

```python
plt.figure(figsize=(5, 4))
sns.violinplot(x='Department', y='Age', data=data)
plt.title('Distribution of Employees\' Ages Across Different Departments')
plt.xlabel('Department')
plt.ylabel('Age')
plt.xticks(rotation=45)
plt.show()
```

### Correlation Between Number of Training Sessions Attended Last Year and Performance Rating

```python
# Calculate correlation
correlation, _ = pearsonr(data['TrainingTimesLastYear'], data['PerformanceRating'])
print(f'Correlation between number of training sessions attended last year and performance rating: {correlation}')

# Scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(x='TrainingTimesLastYear', y='PerformanceRating', data=data)
plt.title('Number of Training Sessions Attended Last Year vs. Performance Rating')
plt.xlabel('Training Times Last Year')
plt.ylabel('Performance Rating')
plt.show()
```

### Relationship Between Years With Current Manager and Work-Life Balance

```python
# Box plot
plt.figure(figsize=(10, 6))
sns.boxplot(x='WorkLifeBalance', y='YearsWithCurrManager', data=data)
plt.title('Relationship Between Years With Current Manager and Work-Life Balance')
plt.xlabel('Work-Life Balance')
plt.ylabel('Years With Current Manager')
plt.show()

# Group by 'WorkLifeBalance' and calculate the mean years with current manager
average_years_with_manager = data.groupby('WorkLifeBalance')['YearsWithCurrManager'].mean()
print(average_years_with_manager)
```
