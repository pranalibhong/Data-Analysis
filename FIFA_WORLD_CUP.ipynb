```python
plt.figure(figsize=(8, 6))
df1['Year'].value_counts().sort_index().plot(kind='bar', title='Matches Over Years (df1)')
plt.xlabel('Year')
plt.ylabel('Number of Matches')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 6))
df1[['Home Team Goals', 'Away Team Goals']].plot(kind='hist', bins=10, alpha=0.7, title='Home vs. Away Team Goals (df1)')
plt.xlabel('Goals')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
top_countries_attendance = df3.sort_values(by='Attendance', ascending=False).head(10)
plt.bar(top_countries_attendance['Country'], top_countries_attendance['Attendance'])
plt.title('Top Countries by Attendance (df3)')
plt.xlabel('Country')
plt.ylabel('Attendance')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 6))
df2['Position'].value_counts().plot(kind='pie', autopct='%1.1f%%', title='Player Positions (df2)')
plt.ylabel('')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
top_countries_wins = df3[['Country', 'Winner']].groupby('Country').count().sort_values(by='Winner', ascending=False).head(10)
plt.bar(top_countries_wins.index, top_countries_wins['Winner'])
plt.title('Countries with Most Wins (df3)')
plt.xlabel('Country')
plt.ylabel('Wins')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 6))
df3.groupby('Year')['GoalsScored'].sum().plot(kind='line', marker='o', title='Goals Scored Over Years (df3)')
plt.xlabel('Year')
plt.ylabel('Goals')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
coach_distribution = df2[['Coach Name', 'Team Initials']].drop_duplicates()['Coach Name'].value_counts().head(10)
plt.bar(coach_distribution.index, coach_distribution)
plt.title('Coach Distribution by Team (df2)')
plt.xlabel('Coach Name')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
referee_distribution = df1['Referee'].value_counts().head(10)
plt.bar(referee_distribution.index, referee_distribution)
plt.title('Referee Distribution (df1)')
plt.xlabel('Referee')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(df3['MatchesPlayed'], df3['GoalsScored'])
plt.title('Matches Played vs. Goals Scored (df3)')
plt.xlabel('Matches Played')
plt.ylabel('Goals Scored')
plt.tight_layout()
plt.show()
```
