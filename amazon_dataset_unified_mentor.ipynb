```python
# Count the frequency of each gender category
sales_counts = df['Sales Channel'].value_counts()

# Create a bar chart
plt.figure(figsize=(8, 4))
ax = sales_counts.plot(kind='bar', color='teal')

# Add data labels to the bars
for i, v in enumerate(sales_counts):
    ax.text(i,v+1, str(v), ha='center')

# Add labels and title
plt.xlabel('sales')
plt.ylabel('Frequency')
plt.title('Bar Chart of Sales Distribution')
plt.xticks(rotation=0)  # Rotate x-axis labels if needed
plt.show()
```

```python
# Convert order_date column to datetime
df['Order Date'] = pd.to_datetime(df['Order Date'])

# Aggregate sales data by the desired time period (e.g., monthly)
# Here, we assume the sales amount column is named 'sales_amount'
monthly_sales = df.resample('M', on='Order Date').sum()

# Plot total sales over time
plt.figure(figsize=(10, 5))
plt.plot(monthly_sales.index, monthly_sales['Total Profit'], marker='o', linestyle='-')
plt.title('Monthly Profit ')
plt.xlabel('Date')
plt.ylabel('Total Profit')
plt.grid(True)
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])

# Group by region and sum units sold
units_sold_by_region = df.groupby('Region')['Units Sold'].sum().reset_index()

# Print the result
print(units_sold_by_region)

# Plot units sold by region
plt.figure(figsize=(10, 5))
plt.bar(units_sold_by_region['Region'], units_sold_by_region['Units Sold'], color='skyblue')
plt.title('Units Sold by Region')
plt.xlabel('Region')
plt.ylabel('Total Units Sold')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
# Calculate the distribution of order priorities
order_priority_distribution = df['Order Priority'].value_counts()

# Print the result
print(order_priority_distribution)

# Plot the distribution using a pie chart
plt.figure(figsize=(6, 6))
plt.pie(order_priority_distribution, labels=order_priority_distribution.index, autopct='%1.1f%%', startangle=140)
plt.title('Order Priority Distribution')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```
