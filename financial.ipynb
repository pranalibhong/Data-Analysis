```python
import matplotlib.pyplot as plt

# Visualize the distribution of market capitalization
plt.hist(data['Mar Cap - Crore'], bins=20, edgecolor='k')
plt.title('Distribution of Market Capitalization')
plt.xlabel('Market Capitalization (Crore)')
plt.ylabel('Frequency')
plt.show()
```

```python
# Line plot of market capitalization over entries
plt.plot(data['S.No.'], data['Mar Cap - Crore'])
plt.title('Market Capitalization Over Entries')
plt.xlabel('S.No.')
plt.ylabel('Market Capitalization (Crore)')
plt.show()
```

```python
import seaborn as sns

# Scatter plot of market capitalization vs quarterly sales
sns.scatterplot(x='Mar Cap - Crore', y='Sales Qtr - Crore', data=data)
plt.title('Market Capitalization vs Quarterly Sales')
plt.xlabel('Market Capitalization (Crore)')
plt.ylabel('Quarterly Sales (Crore)')
plt.show()
```

```python
# Bar chart to compare quarterly sales of different entries
plt.figure(figsize=(10, 6))
plt.bar(data['S.No.'], data['Sales Qtr - Crore'])
plt.title('Quarterly Sales of Different Entries')
plt.xlabel('S.No.')
plt.ylabel('Quarterly Sales (Crore)')
plt.xticks(rotation=45)
plt.show()
```

```python
# Pair plot to show relationships between market capitalization and quarterly sales
sns.pairplot(data[['Mar Cap - Crore', 'Sales Qtr - Crore']])
plt.suptitle('Pair Plot of Market Capitalization and Quarterly Sales')
plt.show()
```
